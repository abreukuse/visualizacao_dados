```
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 1000)
```

```
# dados com as transições de alunos por diferentes IES de caxias do sul
migracoes = pd.read_csv('fluxo_estudantes_2018.csv')
migracoes.head(10)
```

```
siglas = np.load('dicionario_ies_siglas.npy', allow_pickle=True).item()
siglas
```

```
cursos_faltantes = np.load('dicionario_cursos_faltantes.npy', allow_pickle=True).item()
cursos_faltantes
```

```
# alguns cursos estão com um código numério
# é preciso substituir pelo nome.

for cada in ['curso_anterior', 'curso_atual']:
  migracoes[cada].replace(cursos_faltantes, inplace = True)
```

```
# dados de alunos que sairam da ucs e foram pra fsg
ucs = (migracoes.ies_anterior == 'Universidade De Caxias Do Sul')
fsg = (migracoes.ies_atual == 'Centro Universitário Da Serra Gaúcha')

ucs_fsg = migracoes.loc[ucs & fsg,:].copy()
ucs_fsg.head(10)
```

```
migracao_ucs_fsg = ucs_fsg.groupby(['curso_anterior', 'curso_atual']).size()
migracao_ucs_fsg
```

```
df_ucs_fsg = pd.DataFrame(list(migracao_ucs_fsg.keys()), columns = ['UCS', 'FSG'])
df_ucs_fsg['Quantidade'] = migracao_ucs_fsg.values
df_ucs_fsg
```

```
# csv
df_ucs_fsg.to_csv('ucs_para_fsg_cursos.csv', index=False)
```
