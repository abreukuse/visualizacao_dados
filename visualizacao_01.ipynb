```
import pandas as pd
import numpy as np
```

```
nomes_dados = ['alunos_caxias_{}'.format(x) for x in range(2009, 2019)]
nomes_dados
```

```
# importando os csv´s
for nome in nomes_dados:
  vars()[nome] = pd.read_csv(nome + '.csv')
```

```
files = [alunos_caxias_2009,
         alunos_caxias_2010,
         alunos_caxias_2011,
         alunos_caxias_2012,
         alunos_caxias_2013,
         alunos_caxias_2014,
         alunos_caxias_2015,
         alunos_caxias_2016,
         alunos_caxias_2017,
         alunos_caxias_2018]
```

```
situacao = (alunos_caxias_2009.TP_SITUACAO == 'Cursando')
nivel = (alunos_caxias_2009.TP_NIVEL_ACADEMICO == 'Graduação')

cursando = alunos_caxias_2009.loc[situacao & nivel, ['CO_IES']]
contagem = cursando.CO_IES.value_counts()
df = pd.DataFrame(zip(contagem.keys(), contagem.values), columns = ['IES', '2009'])
df
```

```
anos = list(range(2009,2019))
alunos_cursando = ['cursando_{}'.format(x) for x in anos]
alunos_cursando
```

```
for ano, alunos, file in zip(anos, alunos_cursando, files):
  situacao = (file.TP_SITUACAO == 'Cursando')
  nivel = (file.TP_NIVEL_ACADEMICO == 'Graduação')

  cursando = file.loc[situacao & nivel, ['CO_IES']]
  contagem = cursando.CO_IES.value_counts()
  vars()[alunos] = pd.DataFrame(zip(contagem.keys(), contagem.values), columns = ['IES', ano])
```

```
dataframes = [cursando_2009,
              cursando_2010,
              cursando_2011,
              cursando_2012,
              cursando_2013,
              cursando_2014,
              cursando_2015,
              cursando_2016,
              cursando_2017,
              cursando_2018]
```

```
dicionario_ies_siglas = np.load('dicionario_ies_siglas.npy', allow_pickle = True).item()
dicionario_ies_siglas
```

```
for df in dataframes:
  df.IES.replace(dicionario_ies_siglas, inplace = True)
```

```
from functools import reduce
```

```
alunos_cursando_por_ano = reduce(lambda left, right: pd.merge(left, right, how = 'outer', on = 'IES'), dataframes)
alunos_cursando_por_ano
```

```
alunos_cursando_por_ano.to_csv('visualizacao_01.csv', index = False)
```
