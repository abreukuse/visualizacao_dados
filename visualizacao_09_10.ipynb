```
# selecionar só alunos da ucs que estavam em um curso presencial entre 2015 e 2018

anos = (alunos_caxias.NU_ANO_INGRESSO.isin([2015,2016,2017,2018])) # só interessa registros recentes
situacao = (alunos_caxias.TP_SITUACAO.isin([3, 4, 5]))
ucs = (alunos_caxias.CO_IES == 'UNIVERSIDADE DE CAXIAS DO SUL')
presencial = (alunos_caxias.TP_MODALIDADE_ENSINO == 1)

nao_cursando = alunos_caxias.loc[anos & situacao & ucs & presencial,:].copy()
nao_cursando.head() 
```

```
def renomear_colunas(data, mapa):
  '''
  renomear as colunas para montar o antes e o 
  depois dos alunos
  
  argumentos:
    data -> pandas dataframe
    mapa -> dicionário
  '''
  colunas = ['ID_ALUNO',
             'CO_IES',
             'CO_CURSO',
             'TP_SITUACAO',
             'NU_ANO_INGRESSO',
             'TP_GRAU_ACADEMICO',
             'TP_MODALIDADE_ENSINO']
  data = data[colunas].copy()
  data.rename(columns = mapa, inplace = True)
  return data
```

```
colunas = ['CO_IES',
           'CO_CURSO',
           'TP_SITUACAO',
           'NU_ANO_INGRESSO',
           'TP_GRAU_ACADEMICO',
           'TP_MODALIDADE_ENSINO']

novas_nao_cursando = ['ies_anterior',
                      'curso_anterior',
                      'situacao_anterior',
                      'ano_anterior',
                      'grau_anterior',
                      'modalidade_anterior']
novas_cursando = ['ies_atual',
                  'curso_atual',
                  'situacao_atual',
                  'ano_atual',
                  'grau_atual',
                  'modalidade_atual']

mapa_nao_cursando = dict(zip(colunas, novas_nao_cursando))
mapa_cursando = dict(zip(colunas, novas_cursando))

mapa_cursando, mapa_nao_cursando
```

```
antes_presencial = renomear_colunas(nao_cursando, mapa_nao_cursando)
antes_presencial.head()
```

```
depois_ead = renomear_colunas(ead_cursando, mapa_cursando)
depois_ead.head()
```

```
migracao_presencial_ead = antes_presencial.merge(depois_ead, on = 'ID_ALUNO')
migracao_presencial_ead.head(10)
```

```
# só troca os códigos pelos nomes
genero = {2: 'Masculino',
          1: 'Feminino'}

situacao = {2: 'Cursando',
3: 'Matrícula trancada',
4: 'Desvinculado do curso',
5: 'Transferido para outro curso da mesma IES',
6: 'Formado',
7: 'Falecido'}

grau = {1: 'Bacharelado',
        2: 'Licenciatura',
        3: 'Tecnólogo',
        4: 'Bacharelado e Licenciatura'}

modalidade = {1: 'Presencial',
              2: 'Curso a distância'}

nivel = {1: 'Graduação',
         2: 'Sequencial de Formação'}

# formatações
migracao_presencial_ead.situacao_anterior.replace(situacao, inplace=True)
migracao_presencial_ead.grau_anterior.replace(grau, inplace=True)
migracao_presencial_ead.modalidade_anterior.replace(modalidade, inplace=True)
migracao_presencial_ead.situacao_atual.replace(situacao, inplace=True)
migracao_presencial_ead.grau_atual.replace(grau, inplace=True)
migracao_presencial_ead.modalidade_atual.replace(modalidade, inplace = True)

colunas = ['ies_anterior','curso_anterior','ies_atual','curso_atual']

for coluna in colunas:
  migracao_presencial_ead[coluna] = migracao_presencial_ead[coluna].map(lambda x: x.title())

migracao_presencial_ead.head(8)
```
