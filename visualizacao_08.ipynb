```
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 1000)
```

```
alunos_caxias = pd.read_csv('alunos_caxias_2018.csv')
alunos_caxias.head()
```

```
faltantes = np.load('dicionario_cursos_faltantes.npy', allow_pickle=True).item()
```

```
alunos_caxias.NO_CURSO.replace(faltantes, inplace = True)
```

```
# selecionar os registros com a situação 'Cursando' e que começaram na IES em 2018
situacao = alunos_caxias.TP_SITUACAO == 'Cursando' # situação cursando
nivel_academico = (alunos_caxias.TP_NIVEL_ACADEMICO == 'Graduação') #nível de graduação presencial e a distância 
ano = (alunos_caxias.ANO_INGRESSO == 2018)
cursando = alunos_caxias.loc[situacao & nivel_academico & ano,:].copy() # alunos cursando em 2018
cursando.head()
```

```
# para gerar a cross table abaixo

ies_curso = cursando[['CO_IES','NO_CURSO','TP_MODALIDADE_ENSINO']].copy()
ies_curso['Curso_Modalidade'] = ies_curso['NO_CURSO'] + ' - ' + ies_curso['TP_MODALIDADE_ENSINO']
ies_curso.drop(['NO_CURSO', 'TP_MODALIDADE_ENSINO'], axis = 1, inplace = True)

ies_curso.head()
```

```
siglas = np.load('dicionario_ies_siglas.npy', allow_pickle=True).item()
siglas
```

```
tabela_cruzada = pd.crosstab(index = ies_curso.Curso_Modalidade, columns = ies_curso.CO_IES)
tabela_cruzada.rename(siglas, axis = 1, inplace = True)
tabela_cruzada
```

```
# csv
tabela_cruzada.to_csv('alunos_por_curso_ies_e_modalidade.csv')
```
