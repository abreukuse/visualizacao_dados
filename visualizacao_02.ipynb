```
import pandas as pd
import numpy as np
```

```
nomes_dados = ['alunos_caxias_{}'.format(x) for x in range(2009, 2019)]
nomes_dados
```

```
# importando os csv´s
for nome in nomes_dados:
  vars()[nome] = pd.read_csv(nome + '.csv')
```

```
files = [alunos_caxias_2009,
         alunos_caxias_2010,
         alunos_caxias_2011,
         alunos_caxias_2012,
         alunos_caxias_2013,
         alunos_caxias_2014,
         alunos_caxias_2015,
         alunos_caxias_2016,
         alunos_caxias_2017,
         alunos_caxias_2018]

anos = list(range(2009,2019))
```

```
situacao = (alunos_caxias_2009.TP_SITUACAO == 'Cursando')
nivel = (alunos_caxias_2009.TP_NIVEL_ACADEMICO == 'Graduação')
ano = (alunos_caxias_2009.ANO_INGRESSO == 2009)

cursando = alunos_caxias_2009.loc[situacao & nivel & ano, ['CO_IES']]
contagem = cursando.CO_IES.value_counts()
df = pd.DataFrame(zip(contagem.keys(), contagem.values), columns = ['IES', '2009'])
df
```

```
alunos_ingressantes = ['ingressantes_{}'.format(x) for x in anos]
alunos_ingressantes
```

```
for ano, alunos, file in zip(anos, alunos_ingressantes, files):
  situacao = (file.TP_SITUACAO == 'Cursando')
  nivel = (file.TP_NIVEL_ACADEMICO == 'Graduação')
  ano_ingresso = (file.ANO_INGRESSO == ano)
  
  cursando = file.loc[situacao & nivel & ano_ingresso, ['CO_IES']]
  contagem = cursando.CO_IES.value_counts()
  vars()[alunos] = pd.DataFrame(zip(contagem.keys(), contagem.values), columns = ['IES', ano])
```

```
dataframes = [ingressantes_2009,
               ingressantes_2010,
               ingressantes_2011,
               ingressantes_2012,
               ingressantes_2013,
               ingressantes_2014,
               ingressantes_2015,
               ingressantes_2016,
               ingressantes_2017,
               ingressantes_2018]
```

```
dicionario_ies_siglas = np.load('dicionario_ies_siglas.npy', allow_pickle = True).item()
dicionario_ies_siglas
```

```
for df in dataframes:
  df.IES.replace(dicionario_ies_siglas, inplace = True)
```

```
from functools import reduce
```

```
alunos_ingressantes_por_ano = reduce(lambda left, right: pd.merge(left, right, how = 'outer', on = 'IES'), dataframes)
alunos_ingressantes_por_ano
```

```
alunos_ingressantes_por_ano.to_csv('visualizacao_02.csv', index = False)
```
