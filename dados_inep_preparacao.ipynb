```
# Essa parte é para pegar os códigos das IES de Caxias e
# associar esses códigos aos nomes das IES.
# A UCS tem o código 13 por exemplo

codigo_ibge_caxias = 4305108 # código do IBGE de Caxias
codigos_ies_caxias = []
nomes_ies_caxias = []

for each in path_ies:
  dummy = pd.read_csv(each, delimiter = '|', encoding = 'latin1')
  for coluna in dummy.columns:
    comparacao = 'CO_MUNICIPIO'
    if fuzz.partial_ratio(comparacao, coluna) >= 95:
      dummy.rename({coluna: comparacao}, axis = 1, inplace = True)
  caxias = (dummy[comparacao] == codigo_ibge_caxias)
  ies_caxias = dummy.loc[caxias,:].copy()
  
  codigos = ies_caxias.CO_IES
  nomes = ies_caxias.NO_IES
  
  nomes_ies_caxias.append(dict(zip(codigos, nomes))) # gerando uma lista de dicionários com os códigos e nomes das IES 
  
  codigos_ies_caxias.append(codigos.unique())
```

```
nomes_arquivos = ['alunos_caxias_{}'.format(x) for x in range(2009,2019)]
nomes_arquivos
```

```
# Aqui é feita a importação dos dados dos alunos.
# Como são arquivos csv muito grandes, entre 3 e 5 GB,
# é preciso abrir os arquivos em pedaços para não usar toda a memória.
# Para fazer isso, o parametro chunksize é usado em pd.read_csv.
# DEMORA CERCA DE 15 MINUTOS PARA RODAR

for linhas, colunas, alunos, nome in zip(codigos_ies_caxias, lista_colunas, path_alunos, nomes_arquivos):
  chunks = pd.read_csv(alunos, 
                     delimiter = '|', encoding = 'latin1', chunksize = 500_000)
  
  dataframes = []

  for chunk in chunks:
    filtro = chunk.loc[chunk.CO_IES.isin(linhas), colunas].copy()
    dataframes.append(filtro)

  vars()[nome] = pd.concat(dataframes)
```

```
# dicionários para fazer as mudanças dos códigos para o que eles significam nos dados de 2009, 2010, 2011, 2017 e 2018 

genero_2009 = {0: 'Masculino',
              1: 'Feminino'}

genero_2017_2018 = {2: 'Masculino',
                    1: 'Feminino'}

situacao = {1: 'Provável Formando', # essa descrição só está nos dados de 2009
            2: 'Cursando',
            3: 'Matrícula trancada',
            4: 'Desvinculado do curso',
            5: 'Transferido para outro curso da mesma IES',
            6: 'Formado',
            7: 'Falecido'}

grau = {1: 'Bacharelado',
        2: 'Licenciatura',
        3: 'Tecnólogo',
        4: 'Bacharelado e Licenciatura'} # só no dados de 2009

modalidade = {1: 'Presencial',
              2: 'a distância'}

nivel = {1: 'Graduação',
         2: 'Sequencial de Formação'}
```

```
# formatações 2009, 2010 e 2011

dados = [alunos_caxias_2009, alunos_caxias_2010, alunos_caxias_2011]

cols = ['TP_SEXO','TP_SITUACAO','TP_GRAU_ACADEMICO', 'TP_MODALIDADE_ENSINO', 'TP_NIVEL_ACADEMICO']
dicis = [genero_2009, situacao, grau, modalidade, nivel]


for data in dados:  
  for col, dic in zip(cols, dicis):
    data[col].replace(dic, inplace = True)
```

```
# formatações 2017 e 2018

dados = [alunos_caxias_2017, alunos_caxias_2018]

cols = ['TP_SEXO','TP_SITUACAO','TP_GRAU_ACADEMICO', 'TP_MODALIDADE_ENSINO', 'TP_NIVEL_ACADEMICO']
dicis = [genero_2017_2018, situacao, grau, modalidade, nivel]


for data in dados:  
  for col, dic in zip(cols, dicis):
    data[col].replace(dic, inplace = True)
    
# os outros anos já estão no formato desejado
```

```
# Ainda é preciso colcar os nomes das IES para todos os anos
# e colocar os nomes dos cursos.
# Esses dicionarios foram criados usando os dados dos cursos e também com o site do e-MEC para 
# pegar os nomes dos cursos em que só tinham os códigos.

dicionario_cursos = np.load('dicionario_cursos.npy', allow_pickle = True).item()
dicionario_cursos_faltantes = np.load('dicionario_cursos_faltantes.npy', allow_pickle = True).item()
```

```
# colocar os nomes das IES em todos os dados
for nomes, df in zip(nomes_ies_caxias, dataframes):
  df['CO_IES'].replace(nomes, inplace = True)
```

```
# colocar os nomes dos cursos para os anos de 2017 e 2018

for df in [alunos_caxias_2017, alunos_caxias_2018]:
  for dici in [dicionario_cursos, dicionario_cursos_faltantes]:
    df['CO_CURSO'].replace(dici, inplace = True)  
```
